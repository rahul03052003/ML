```python
import numpy as np
def sig(a,p):
    return np.mean((a-p)**2)

a_v=np.random.randint(1,121,5)
p_v=np.random.randint(1,121,5)

sig(a_v,p_v)
```

```python
def sigmiod(x):
    return 1/(1+np.exp(-x))
n=np.arange(1,11)
print(sigmiod(n))
```

```python
def soft(x):
    exp_x=np.exp(x-np.max(x))
    return exp_x/np.sum(exp_x)

x=np.array([2.0,3,4.0])
print(soft(x))
```

```python
from sklearn.linear_model import LinearRegression
df=pd.read_csv('placement.csv')
df.head()
```

```python
df
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
y_pred=lr.predict(x_test)
print("Mse: ",mean_squared_error(y_test,y_pred))
print("Mae: ",mean_absolute_error(y_test,y_pred))
print("Rmse: ",root_mean_squared_error(y_test,y_pred))
```

```python
df=pd.read_csv('student_clustering_KMeans.csv')
df.head()
```

```python
plt.scatter(df['cgpa'],df['iq'])
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kn=KMeans(n_clusters=i)
    kn.fit_predict(df)
    wcss.append(kn.inertia_)
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
x=df.iloc[:,:]
km=KMeans(n_clusters=4)
y_means=km.fit_predict(x)
y_means

```

```python
plt.scatter(x.iloc[y_means==0,0],x.iloc[y_means==0,1],color='red')
plt.scatter(x.iloc[y_means==1,0],x.iloc[y_means==1,1],color='red')
plt.scatter(x.iloc[y_means==2,0],x.iloc[y_means==2,1],color='red')
plt.scatter(x.iloc[y_means==3,0],x.iloc[y_means==3,1],color='red')
```

```python
from sklearn.metrics import silhouette_score
print(silhouette_score(x,y_means))
```

```python
df=pd.read_csv('insurance_data.csv')
df.head()
```

```python
x_test
```

```python
from  sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score
from sklearn import datasets
iris=datasets.load_iris()
iris
```

```python
x=iris.data[:,:2]
y=iris.target
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
dt=DecisionTreeClassifier(criterion='gini',random_state=4)
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
y_pred
```

```python
acc=accuracy_score(y_test,y_pred)
print(100*acc)
```

```python
plt.figure(figsize=(12,8))
plot_tree(dt,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
plt.show()
```

```python
dtt=DecisionTreeClassifier(criterion='gini',max_depth=3,max_features=4,max_leaf_nodes=3,min_impurity_decrease=0.2,min_samples_leaf=3,min_samples_split=4,random_state=4)

dtt.fit(x_train,y_train)
```

```python
y_pred=dtt.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
acc*100
```

```python
plt.figure(figsize=(12,8))
plot_tree(dtt,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```
